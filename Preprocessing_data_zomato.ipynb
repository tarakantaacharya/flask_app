```python
import pandas as pd

# Step 1: Concatenate only the "R.res_id" column from df1 to df5
combined_df = pd.concat([df1[["R.res_id"]], df2[["R.res_id"]], df3[["R.res_id"]], df4[["R.res_id"]], df5[["R.res_id"]]])

# Step 2: Remove duplicates
unique_ids = combined_df["R.res_id"].drop_duplicates()

# Step 3: Count matching and unmatching restaurant IDs
matching_count = df["Restaurant ID"].isin(unique_ids).sum()
unmatching_count = len(df) - matching_count

# Step 4: Print the results
print(f"Matching count: {matching_count}")
print(f"Unmatching count: {unmatching_count}")
```

```python
import pandas as pd

# Extract only 'Restaurant ID' column from df (keeping all its other columns)
df_restaurant_ids = df[['Restaurant ID']].drop_duplicates()

# Find common columns among df1 to df5 (ensuring R.res_id is always included)
common_columns = set(df1.columns) & set(df2.columns) & set(df3.columns) & set(df4.columns) & set(df5.columns)
common_columns.add("R.res_id")  # Ensure 'R.res_id' is included

# Extract only the common columns from each DataFrame
df1_common = df1[list(common_columns)]
df2_common = df2[list(common_columns)]
df3_common = df3[list(common_columns)]
df4_common = df4[list(common_columns)]
df5_common = df5[list(common_columns)]

# Concatenate all the filtered DataFrames
combined_df = pd.concat([df1_common, df2_common, df3_common, df4_common, df5_common])

# Remove duplicate R.res_id values
unique_df = combined_df.drop_duplicates(subset=["R.res_id"])

# Merge df with unique_df on 'Restaurant ID' and 'R.res_id'
merged_df = df.merge(unique_df, left_on="Restaurant ID", right_on="R.res_id", how="left")

# Count the number of matches and non-matches
matching_count = merged_df["R.res_id"].notna().sum()
unmatching_count = merged_df["R.res_id"].isna().sum()

# Display results
print(f"Total Restaurant IDs in df: {len(df)}")
print(f"Unique Restaurant IDs in df1 to df5: {len(unique_df)}")
print(f"Matching count: {matching_count}")
print(f"Unmatching count: {unmatching_count}")

# Show a sample of the merged DataFrame
print(merged_df.head())
```

```python
import pandas as pd

### 🔹 Step 1: Drop Unwanted Columns (EXCLUDING URL-related columns) ###
columns_to_drop = [
    "apikey", "switch_to_order_menu", "deeplink",  # Example: Dropping unnecessary fields
    "establishment_types", "offers", "id"         # Example: Dropping redundant fields
]

merged_df = merged_df.drop(columns=[col for col in columns_to_drop if col in merged_df.columns], errors='ignore')

### 🔹 Step 2: Rename Columns (Keeping URL-related Columns) ###
rename_columns = {
    "Restaurant Name": "Name",
    "Country Code": "Country_Code",
    "City": "City",
    "Locality": "Locality",
    "Locality Verbose": "Full Address",
    "Cuisines": "Cuisine",
    "Average Cost for two": "Avg Cost for Two",
    "Price range": "Price Range",
    "Aggregate rating": "Rating",
    "Rating text": "Rating Text",
    "Rating color": "Rating Color",
    "Votes": "Total Votes",
}

merged_df = merged_df.rename(columns=rename_columns)

### 🔹 Step 3: Save the Cleaned Data ###
merged_df.to_csv("cleaned_merged_data.csv", index=False)

print("✅ Columns Dropped & Renamed Successfully! (URLs are KEPT)")
print("📌 Final DataFrame Shape:", merged_df.shape)
```
